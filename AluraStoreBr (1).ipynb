### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento Loja 1: R$ {faturamento_loja:.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:.2f}")
```

# 2. Vendas por Categoria


```python
# Loja 1
print("Vendas por Categoria - Loja 1")
print(loja['Categoria do Produto'].value_counts())

# Loja 2
print("\nVendas por Categoria - Loja 2")
print(loja2['Categoria do Produto'].value_counts())

# Loja 3
print("\nVendas por Categoria - Loja 3")
print(loja3['Categoria do Produto'].value_counts())

# Loja 4
print("\nVendas por Categoria - Loja 4")
print(loja4['Categoria do Produto'].value_counts())
```

# 3. Média de Avaliação das Lojas

```python
avaliacao_media_loja = loja['Avaliação da compra'].mean()
avaliacao_media_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_media_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_media_loja4 = loja4['Avaliação da compra'].mean()

print("Média de Avaliação - Loja 1:", avaliacao_media_loja)
print("Média de Avaliação - Loja 2:", avaliacao_media_loja2)
print("Média de Avaliação - Loja 3:", avaliacao_media_loja3)
print("Média de Avaliação - Loja 4:", avaliacao_media_loja4)
```

# 4. Produtos Mais e Menos Vendidos

```python
# Mais vendidos
mais_vendidos_loja = loja['Produto'].value_counts().head(5)
mais_vendidos_loja2 = loja2['Produto'].value_counts().head(5)
mais_vendidos_loja3 = loja3['Produto'].value_counts().head(5)
mais_vendidos_loja4 = loja4['Produto'].value_counts().head(5)

print("Top 5 Produtos Mais Vendidos - Loja 1")
print(mais_vendidos_loja)
print("\nTop 5 Produtos Mais Vendidos - Loja 2")
print(mais_vendidos_loja2)
print("\nTop 5 Produtos Mais Vendidos - Loja 3")
print(mais_vendidos_loja3)
print("\nTop 5 Produtos Mais Vendidos - Loja 4")
print(mais_vendidos_loja4)
```

```python
# Menos vendidos
menos_vendidos_loja = loja['Produto'].value_counts().tail(5)
menos_vendidos_loja2 = loja2['Produto'].value_counts().tail(5)
menos_vendidos_loja3 = loja3['Produto'].value_counts().tail(5)
menos_vendidos_loja4 = loja4['Produto'].value_counts().tail(5)

print("\nTop 5 Produtos Menos Vendidos - Loja 1")
print(menos_vendidos_loja)
print("\nTop 5 Produtos Menos Vendidos - Loja 2")
print(menos_vendidos_loja2)
print("\nTop 5 Produtos Menos Vendidos - Loja 3")
print(menos_vendidos_loja3)
print("\nTop 5 Produtos Menos Vendidos - Loja 4")
print(menos_vendidos_loja4)
```

# 5. Frete Médio por Loja


```python
frete_medio_loja = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print(f"Custo médio de frete - Loja 1: R${frete_medio_loja:.2f}")
print(f"Custo médio de frete - Loja 2: R${frete_medio_loja2:.2f}")
print(f"Custo médio de frete - Loja 3: R${frete_medio_loja3:.2f}")
print(f"Custo médio de frete - Loja 4: R${frete_medio_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados de faturamento
faturamentos = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Gráfico de faturamento
plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title('Faturamento por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.show()
```
